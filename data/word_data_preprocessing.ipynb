```python
import pandas as pd
# 학년, 학기 기록하는 함수
def function(grade, data_1, data_2, total_data):
  semester_1_data = pd.read_csv(data_1)
  semester_2_data = pd.read_csv(data_2)
  total_data = pd.read_csv(total_data)

  total_data['학년'] = grade
  total_data['학기'] = 2
  total_data = total_data.reindex(['학년', '학기', '낱말', '빈도수'], axis = 1)

  for word in semester_1_data.iloc[:,0]:
    total_data.loc[(total_data['낱말'] == word), '학기'] = 1

  return total_data
```

```python
first_grade_data = function(1,'kor-1-1.csv','kor-1-2.csv','kor-1.csv') # 1학년 1학기, 2학기
second_grade_data = function(2,'kor-2-1.csv','kor-2-2.csv','kor-2.csv') # 2학년 1학기, 2학기
third_grade_data = function(3,'kor-3-1.csv','kor-3-2.csv','kor-3.csv') # 3학년 1학기, 2학기
firth_grade_data = function(4,'kor-4-1.csv','kor-4-2.csv','kor-4.csv') # 4학년 1학기, 2학기
fifth_grade_data = function(5,'kor-5-1.csv','kor-5-2.csv','kor-5.csv') # 5학년 1학기, 2학기
sixth_grade_data = function(6,'kor-6-1.csv','kor-6-2.csv','kor-6.csv') # 6학년 1학기, 2학기
```

```python
total_data = pd.concat([first_grade_data, second_grade_data, third_grade_data, firth_grade_data, fifth_grade_data, sixth_grade_data], axis=0, ignore_index=True)
```

```python
total_data.sort_values(by=['학년', '학기'], ascending=True)
```
